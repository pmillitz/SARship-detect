```python
# Load the numpy archive
#shoreline_data = np.load('../../data/SARFishSample/GRD/validation/S1B_IW_GRDH_1SDV_20200803T075721_20200803T075746_022756_02B2FF_033A.SAFE' \
#                         '/measurement/S1B_IW_GRDH_1SDV_20200803T075721_20200803T075746_022756_02B2FF_033A_global_shoreline_vector.npy', allow_pickle=True)


shoreline_data = np.load('../../data/SARFishSample/Shoreline/validation/5c3d986db930f848v_shoreline.npy', allow_pickle=True)

# Inspect the data structure
print(f"Data type: {type(shoreline_data)}")
print(f"Shape: {shoreline_data.shape}")
print(f"Data example: {shoreline_data[:5] if shoreline_data.size > 5 else shoreline_data}")
```

```python
import matplotlib.pyplot as plt

# Extract x (longitude) and y (latitude) coordinates
x_coords = [coord[0] for element in shoreline_data for coord in element]
y_coords = [coord[1] for element in shoreline_data for coord in element]

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x_coords, y_coords, linestyle='-', marker='o', markersize=3)

# Labels and title
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Shoreline Plot")
plt.grid(True)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

try:
    # Option 1: Load from a file
    shoreline_data = np.load('../../data/SARFishSample/SLC/validation/S1B_IW_SLC__1SDV_20200803T075720_20200803T075748_022756_02B2FF_E5D2.SAFE' \
                             '/measurement/S1B_IW_SLC__1SDV_20200803T075720_20200803T075748_022756_02B2FF_E5D2_3_xView3_shoreline.npy', allow_pickle=True)
except:
    print("something went wrong - check the filepath!")

# Create a simple but effective plot
plt.figure(figsize=(10, 8))

# Plot the shoreline as a continuous line
plt.plot(shoreline_data[:, 0], shoreline_data[:, 1], 
         color='#1F77B4',  # blue color
         linewidth=0.8,    # Thin line for detailed coastline
         alpha=0.8)        # Slight transparency

# Add a title and labels
plt.title('NW Iceland Shoreline', fontsize=14)
plt.xlabel('X Coordinate (meters)', fontsize=12)
plt.ylabel('Y Coordinate (meters)', fontsize=12)

# Add a grid for better reference
plt.grid(True, alpha=0.3, linestyle='--')

# Set axis limits with a small buffer
buffer = (shoreline_data[:, 0].max() - shoreline_data[:, 0].min()) * 0.05  # 5% buffer
plt.xlim(shoreline_data[:, 0].min() - buffer, shoreline_data[:, 0].max() + buffer)
plt.ylim(shoreline_data[:, 1].min() - buffer, shoreline_data[:, 1].max() + buffer)

# Improve layout
plt.tight_layout()

# Save the figure (optional)
#plt.savefig('iceland_shoreline_outline.png', dpi=150)

# Show the plot
plt.show()
```
